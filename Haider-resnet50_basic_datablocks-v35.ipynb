**Notes on Stratified K-Fold Split vs Stratified Shuffle Split**

helpful links to show what is these types of splits:

https://stackoverflow.com/questions/45969390/difference-between-stratifiedkfold-and-stratifiedshufflesplit-in-sklearn/52677641#52677641

there is discussion here that this implementation is better than scikit multilearn implementation of skmultilearn.modelselection.iterativetraintestsplit  (See my diigo highlights)
https://www.kaggle.com/c/human-protein-atlas-image-classification/discussion/67819

Also it has a discussion on another approach to do this multilable stratified split which is not covered in this library nor in the scikit library which is *labelsets-based stratification* vs our current approach (and scikit): * iterative stratification*

In [Sechidis et al. (2011)](http://lpis.csd.auth.gr/publications/sechidis-ecmlpkdd-2011.pdf), the authors compare random sampling, labelset-based stratification, and iterative stratification (the method I have implemented) for multiple datasets. There are cases where labelsets-stratification is better than iterative stratification. These cases are typically when the ratio of unique labelsets to the number of samples is small. In cases where the ratio of unique labelsets to the number of samples is not small, iterative stratification is often the better choice.


**P.S.:**
There is also *Repeated Stratified K-Fold* Repeats Stratified K-Fold n times with different randomization in each repetition.

Also please note that KFold and also the stratified KFold (multilabel or not) have no split by percentage. Because if your n_split=2 then it will split the data into half train and half test. If n_split=5 it will split it into 80% train and 20% test. Since it tries to focus on the test set to cover all data within those n_split iterations. In the Stratified Shuffle split this still holds true, but there will be shuffling of the data before each splitting iteration, which makes the test set sometimes overlap and will not guarantee that each test set has no same items like  other test splits. See this picture from :https://stackoverflow.com/questions/45969390/difference-between-stratifiedkfold-and-stratifiedshufflesplit-in-sklearn/52677641#52677641

### Do for each CV subset: Train + valid preds, test preds

```python
%%time
import time

preds =[]
preds_t = []
y =[]

for ii in range(len(test_indices)): 
    start_time = time.time()
    print(f"starting training iteration {ii}")
    
    np.random.seed(302)
    src = (ImageItemList.from_df(df, path=path/'train_both', suffix='.png', cols=['Id'])
           .split_by_idxs(train_indices[ii],test_indices[ii])
           .label_from_df(cols=['Target'], sep=' ',  classes=[str(i) for i in range(28)]))

    src.train.x.create_func = open_4_channel
    src.train.x.open = open_4_channel
    src.valid.x.create_func = open_4_channel
    src.valid.x.open = open_4_channel

    test_ids = list(sorted({fname.split('_')[0] for fname in os.listdir(path/'test')}))

    test_fnames = [path/'test'/test_id for test_id in test_ids]

    test_fnames[:5]

    src.add_test(test_fnames, label='0');

    src.test.x.create_func = open_4_channel
    src.test.x.open = open_4_channel

    protein_stats = ([0.08069, 0.05258, 0.05487, 0.08282], [0.13704, 0.10145, 0.15313, 0.13814])

    trn_tfms,_ = get_transforms(do_flip=True, flip_vert=True, max_rotate=30., max_zoom=1,
                          max_lighting=0.05, max_warp=0.)

    # total train+valid dataset = 34956 . while before oversampling it was 31072

    NUM_WORKERS = 8
    BS = 64//4

    data = (src.transform((trn_tfms, _), size=512)
            .databunch(bs=BS, num_workers=NUM_WORKERS).normalize(protein_stats))

    #        .normalize(imagenet_stats)

    def resnet50(pretrained):
        return Resnet4Channel(encoder_depth=50)

    # copied from https://github.com/fastai/fastai/blob/master/fastai/vision/learner.py
    def _resnet_split(m): 
        # import pdb; pdb.set_trace()  ; m[0] returns the body ; m[0][6] returns the 7th block of the body ; m[1] returns the head; 
        # so it seems this function specifies the points that divides the 3 groups for the differential learning rates
        return (m[0][6],m[1])

    f1_score = partial(fbeta, thresh=0.2, beta=1)


    learn = create_cnn(
        data,
        resnet50,
        cut=-2,
        split_on=_resnet_split,
        loss_func=F.binary_cross_entropy_with_logits,
        path=path,    
        metrics=[f1_score], 
    )

    lr = 3e-2/4
    learn.fit_one_cycle(10, slice(lr))
    learn.save(f'stage-1-rn50_sz512-os-from-extrads-cv-{i}')
    learn.unfreeze()
    learn.fit_one_cycle(20, slice(3e-5/4, lr/5))
    learn.save(f'stage-2-rn50_sz512-os-from-extrads-cv-{i}')

    preds[ii],y[ii] = learn.get_preds(DatasetType.Valid)
    # y = to_np(y)
    # preds = np.stack(preds, axis=-1)
    # pred = preds.mean(axis=-1)  # This need when in fastai0.7 used TTA
    # pred = preds
    preds_t[ii],_ = learn.get_preds(DatasetType.Test)
    
    elapsed_time = time.time() - start_time
    print(f"It took {elapsed_time/3600} hours for this iteration")
```

## Iafoss thresholds and other tweaking methods below

```python
# https://www.kaggle.com/iafoss/pretrained-resnet34-with-rgby-0-460-public-lb
# Instead of 0.5, one can adjust the values of the threshold for each class individually to boost the score. 
# The code below does it automatically.

from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
import scipy.optimize as opt

def sigmoid_np(x):
    return 1.0/(1.0 + np.exp(-x))

def F1_soft(preds,targs,th=0.0,d=25.0):
#     preds = to_np(preds)
    preds = sigmoid_np(d*(preds - th))
#     targs = to_np(targs)
#     targs = targs.astype(np.float)
    score = 2.0*(preds*targs).sum(axis=0)/((preds+targs).sum(axis=0) + 1e-6)
    return score

def fit_val(x,y):
    params = np.zeros(len(name_label_dict))
    wd = 1e-5
    error = lambda p: np.concatenate((F1_soft(x,y,p) - 1.0, wd*p), axis=None)
    p, success = opt.leastsq(error, params)
    return p
```

Using CV to prevent overfitting the thresholds:

```python
# This function is trying to minimize (F1-1)^2 + wd^2*th^2. So, you are trying to get F1 close to 1. 
# while if something goes wrong, like there are just a few data points, thresholds do not go far away from 0.0. 
# Also, I'm using a soft version of F1 to have stable gradient in least square minimization.

from sklearn.model_selection import train_test_split
th, score, cv = 0,0,10
for i in range(cv):
    xt,xv,yt,yv = train_test_split(pred,y,test_size=0.5,random_state=i)
    th_i = fit_val(xt,yt)
    th += th_i
    score += f1_score(yv, xv>th_i, average='macro')
th/=cv
score/=cv
print('Thresholds: ',th)
print('F1 macro avr:',score)
print('F1 macro: ',f1_score(y, pred>th, average='macro'))
print('F1 micro: ',f1_score(y, pred>th, average='micro'))
```

```python
def Count_soft(preds,th=0.0,d=50.0):
    preds = sigmoid_np(d*(preds - th))
    return preds.mean(axis=0)

def fit_test(x,y):
    params = np.zeros(len(name_label_dict))
    wd = 1e-5
    error = lambda p: np.concatenate((Count_soft(x,p) - y,
                                      wd*p), axis=None)
    p, success = opt.leastsq(error, params)
    return p
```

```python

from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
import scipy.optimize as opt

# preds_y = model.predict(valid_x) # in his code mine is : preds_y -> preds; valid_y -> y

thresholds = np.linspace(0, 1, 1500)
score = 0.0
best_threshold=0.0
best_val = 0.0
for threshold in thresholds:
    score = f1_score(y > 0.5, preds_ave > threshold, average='macro')
    if score > best_val:
        best_threshold = threshold
        best_val = score
    print("Threshold %0.4f, F1: %0.4f" % (threshold,score))

print("BEST: %0.5f, F1: %0.5f" % (best_threshold,best_val))

```
